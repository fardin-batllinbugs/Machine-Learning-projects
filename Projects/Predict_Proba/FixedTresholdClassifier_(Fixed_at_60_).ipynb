```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import FixedThresholdClassifier



model = LogisticRegression()
model.fit(X_train, y_train)

## setting manually at 60% threshold
manual_threshold_classifier = FixedThresholdClassifier(model, threshold=0.6,response_method="predict_proba")



# 5 new students with different study hours
test_hours = np.array([
    [1.8],
    [2.2],
    [2.8],
    [3.2],
    [4.5]
])

# Get probabilities (to see what the model thinks)
probabilites = model.predict_proba(test_hours)
proba_fail=probabilites[:,0]
proba_pass=probabilites[:,1]
print(proba_fail)


####
# Get predictions using our 60% manual threshold
predictions = manual_threshold_classifier.predict(test_hours)
print(predictions)
```
